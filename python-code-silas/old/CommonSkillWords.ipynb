```python
import pandas as pd

# Specify the columns to load
columns_to_load = ['Title', 'Company', 'Location', 'JobDescription', 
                   'JobRequirment', 'RequiredQual', 
                   'ApplicationP', 'AboutC']

# Load the data from the CSV file
df = pd.read_csv('../datasets/job_postings_1.csv', usecols=columns_to_load)
df
```

```python
# Concatenate each column into a single string
concatenated_series = df.apply(lambda row: ' '.join(row.values.astype(str)), axis=1)

# Convert the Series to a numpy array
arr = concatenated_series.to_numpy()
arr.size
```

```python
# take an array of strings in python, make them all lowercase a-z and space characters only, then could how many times each word ocurs in the total corpus of text and display the results from most frequent to least frequent

import collections
import re

merged_string = ' '.join(arr)
clean_string = re.sub('[^a-z\s]', '', merged_string.lower())
clean_words = clean_string.split()
word_counts = collections.Counter(clean_words)
most_common_words = word_counts.most_common()

for word, count in most_common_words:
    print(word, count)
```

```python
# Modify the existing Python code to use the NLTK library and ensure the final list only displays nouns, verbs, and adjectives. All forms of a word should be considered the same word. word like 'run', 'runs', 'running'

import collections
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from tqdm import tqdm

nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

print('here 1')

merged_string = ' '.join(arr)
clean_string = re.sub('[^a-z\s]', '', merged_string.lower())
tokens = word_tokenize(clean_string)

print('here 2')

lemmatizer = WordNetLemmatizer() 
stop_words = set(stopwords.words('english')) 

print('here 3')

tagged_tokens = pos_tag(tokens)
filtered_words = []

print('here 4')

for word, tag in tqdm(tagged_tokens):
    if tag in ['NN', 'NNS', 'NNP', 'NNPS', 'VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ', 'JJ', 'JJR', 'JJS']:
        lemmatized_word = lemmatizer.lemmatize(word)
        if lemmatized_word not in stop_words:
            filtered_words.append(lemmatized_word)

word_counts = collections.Counter(filtered_words)
most_common_words = word_counts.most_common()

for word, count in tqdm(most_common_words):
    print(word, count)
```
